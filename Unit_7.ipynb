```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(open('flipkart.html','r',encoding='utf-8'))
print(soup.prettify())
```

```python

Line_tag = soup.find_all('div',class_='KzDlHZ')
Rate_tag = soup.find_all('div',class_='XQDdHH')
Price_tag = soup.find_all('div',class_='Nx9bqj _4b5DiR')
```

```python
Names=[]
colors=[]
ROMs=[]
for i in Line_tag:
    Names.append(i.text.strip().split(" (")[0])
    colors.append(i.text.strip().split(" (")[1].split(",")[0])
    ROMs.append(i.text.strip().split(" (")[1].split(", ")[1].split()[0])
print(Names)
print(colors)
print(ROMs)
```

```python
Ram_tag_box = soup.find_all('ul',class_='G4BRas')
Rams = []
for i in Ram_tag_box:
    Rams.append(i.find('li').text.strip().split('|')[0].split(' GB RAM')[0])
    
ram_1=[]    
ram=soup.find_all('li',class_='J+igdf')   
# sir 
for i in ram:
    if "RAM" in i.text:
        ram_1.append(i.text.split('|')[0][:-4])

print(Rams)
print(ram_1)
```

```python
Prices=[]
price=soup.find_all('div',class_='Nx9bqi _4b5DiR')
for i in Price_tag:
    Prices.append(i.text.strip())
print(Prices)    
```

```python
import requests,json
def get_coordinates():
    api_key='ad22df9b3d3f927ac2cd29f832c099b3'
    city=input('Enter City Name')
    #request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    request_url='http://api.openweathermap.org/geo/1.0/direct?q='+city+'&appid='+api_key
    response=requests.get(request_url)
    coordinates=response.json()
    print(json.dumps(coordinates,indent=5))
    lattitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    return lattitude,longitude,city
lat,lon,city=get_coordinates()        
print(city,"Lattitude & Longitude")
print("- - - - - - - - - - - - - - - ")
print("Lattitude:",lat)
print("Longitude:",lon)
```

```python
import requests,json
def forcast():
    api_key='ad22df9b3d3f927ac2cd29f832c099b3'
    request_url=f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    weather_data=response.json()
    print(json.dumps(weather_data,indent=5))
    return weather_data
weather_data=forcast()
temperture=[]
pressure=[]
humidity=[]
wind_speed=[]
visibility=[]
date=[]
time=[]
des=[]
#print(lis)
for i in weather_data['list']:
        temperture.append(i['main']['temp'])
        pressure.append(i['main']['pressure'])
        humidity.append(i['main']['humidity'])
        wind_speed.append(i['wind']['speed'])
        visibility.append(i['visibility'])
        des.append(i['weather'][0]["description"])
        date.append(i["dt_txt"].split(" ")[0])  
        time.append(i["dt_txt"].split(" ")[1])  
print(temperture)
print(pressure)
print(humidity)
print(wind_speed)
print(visibility)
print(date)    
print(time)
print(des)
```
